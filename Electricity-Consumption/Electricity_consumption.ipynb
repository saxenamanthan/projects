```python
# Exploratory Data Analysis via addressing the Stakeholder Questions

# Q1. Calculate the Electricity bill for the shared observations
```

```python
df1 = pd.read_csv('/content/Tariff_rates.csv')
```

```python
df1.head()
```

```python
electricity_bill = pd.merge(df,df1,how = 'inner', on = 'City')
```

```python
electricity_bill
```

```python
electricity_bill['TariffRate'].nunique()
```

```python
electricity_bill.isna().sum()
```

```python
electricity_bill['Bill']=electricity_bill['Total Monthly Hours']*electricity_bill['TariffRate']
```

```python
electricity_bill
```

```python
# Q2 Calculate the average monthly hours of the usage for the cities in Descending order
```

```python
#groupby never works without aggregation

electricity_bill.groupby('City')['Total Monthly Hours'].mean().sort_values(ascending=False)
```

```python
# Visualizing the solution for the Q2

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,9), dpi = 300)
plt.bar(electricity_bill.groupby('City')['Total Monthly Hours'].mean().sort_values(ascending=False))
```

```python
plt.figure(figsize=(20,9), dpi = 300)

sns.barplot(electricity_bill.groupby('City')['Total Monthly Hours'].mean().sort_values(ascending=False))
```

```python
# Q3. What are the top 3 companies in terms of average monthly hours of electricity usage

electricity_bill.groupby('Company')['Total Monthly Hours'].mean().sort_values(ascending=False).head(3)
```

```python
sns.barplot(electricity_bill.groupby('Company')['Total Monthly Hours'].mean().sort_values(ascending=False).head(3))
```
